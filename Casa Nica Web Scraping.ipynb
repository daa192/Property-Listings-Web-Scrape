```python
from __future__ import print_function
import os.path
from collections import defaultdict
import string
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'http://www.casanica.com/venta-de-terrenos-en-nicaragua-l0-6-0.html'
pages = str(list(range(2,70)))
```

```python
response = get('http://www.casanica.com/venta-de-terrenos-en-nicaragua-l0-6-0.html')
soup = BeautifulSoup(response.text,'html.parser')
containers = soup.find_all('div', class_ = 'anunciondii')
```

```python
# Use a defaultdict with an empty list because it eases the DataFrame creation
names = []
desc = []
location = []
price = []
```

```python
#loop for first page
for container in containers:
            # The name
            name = container.h2.a.text
            names.append(name)

            # description
            desc1 = container.p.text
            desc.append(desc1)
            
            #location
            loc1 = container.h3.text
            location.append(loc1)
        
            #price
            price1 = container.find('div', class_ = 'anuncio_priceii')
            price1 = container.text
            price.append(price1)
```

```python
for page in pages:
    
    # Make a get request
    response = get('http://www.casanica.com/venta-de-terrenos-en-nicaragua-l0-6-0-'+page+'.html')
    soup = BeautifulSoup(response.text,'html.parser')
    containers = soup.find_all('div', class_ = 'anunciondii')

    # Extract data from individual listing container
    for container in containers:
            # The name
            name = container.h2.a.text
            names.append(name)

            # description
            desc1 = container.p.text
            desc.append(desc1)
            
            #location 
            loc1 = container.h3.text
            location.append(loc1)
        
            #price
            price1 = container.find('div', class_ = 'anuncio_priceii')
            price1 = container.text
            price.append(price1)
```

```python
import pandas as pd

test_df = pd.DataFrame({'Nombre': names,
                       'Descripcion': desc,
                       'Lugar': location,
                       'Precio': price})
print(test_df.info())
test_df
```
