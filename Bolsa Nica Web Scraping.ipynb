```python
from __future__ import print_function
import os.path
from collections import defaultdict
import string
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'http://www.bolsanica.com/V/-1-/venta-de-terrenos-en-nicaragua-cat-11-.html'
pages = ['','p2','p3','p4','p5']
```

```python
# Use a defaultdict with an empty list because it eases the DataFrame creation
names = []
desc = []
size = []
price = []
```

```python
# For every page in the website
for page in pages:
    
    # Make a get request
    response = get('http://www.bolsanica.com/V/-1-/venta-de-terrenos-en-nicaragua-cat-11-'+page+'.html')
    soup = BeautifulSoup(response.text,'html.parser')
    containers = soup.find_all('div', class_ = 'item')

    # Extract data from individual listing container
    for container in containers:
            # The name
            name = container.h3.a.text
            names.append(name)

            # description
            desc1 = container.p.text
            desc.append(desc1)
        
            #price
            price1 = container.find('div', class_ = 'price')
            price1 = price1.text
            price.append(price1)
```

```python
import pandas as pd

test_df = pd.DataFrame({'Nombre': names,
                       'Descripcion': desc,
                       'Precio': price})
print(test_df.info())
test_df
```
